```python
import os
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
import matplotlib.pyplot as plt
import seaborn as sns

# Directory to save results
output_dir = r"C:/Users/Monet/Desktop/TCPCyberIDS/UNR-IDD/report/"
os.makedirs(output_dir, exist_ok=True)

# Define classifiers
classifiers = {
    'Random Forest': RandomForestClassifier(),
    'MLP': MLPClassifier(max_iter=1000),
    'SVM': SVC(),
    'Bagging Classifier': BaggingClassifier(),
    'KNN': KNeighborsClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'XGBoost': XGBClassifier(),
    'CatBoost': CatBoostClassifier(),
    'Gaussian Naive Bayes': GaussianNB(),
    'Logistic Regression': LogisticRegression(),
    'Gaussian Process': GaussianProcessClassifier()
}

# List of labels for classification
labels = ['Blackhole', 'Diversion', 'Normal', 'Overflow', 'PortScan', 'TCP-SYN']

# Store results for validation
results = []

for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_val)
    
    accuracy = accuracy_score(y_val, y_pred)
    mean_precision = np.mean(precision_score(y_val, y_pred, average=None, zero_division=0))
    mean_recall = np.mean(recall_score(y_val, y_pred, average=None, zero_division=0))
    mean_f1 = np.mean(f1_score(y_val, y_pred, average=None, zero_division=0))
    min_f1 = np.min(f1_score(y_val, y_pred, average=None, zero_division=0))
    
    precision_per_class = precision_score(y_val, y_pred, average=None, zero_division=0)
    recall_per_class = recall_score(y_val, y_pred, average=None, zero_division=0)
    f1_per_class = f1_score(y_val, y_pred, average=None, zero_division=0)
    
    results.append({
        'Model': name,
        'Validation Accuracy': accuracy,
        'Validation Mean Precision': mean_precision,
        'Validation Mean Recall': mean_recall,
        'Validation Mean F1 Score': mean_f1,
        'Validation Minimum F1 Score': min_f1,
        'Validation Precision Per Class': precision_per_class.tolist(),
        'Validation Recall Per Class': recall_per_class.tolist(),
        'Validation F1 Score Per Class': f1_per_class.tolist()
    })

# Save validation results
results_df = pd.DataFrame(results)
results_file = os.path.join(output_dir, 'ablation_model_Validation_results.csv')
results_df.to_csv(results_file, index=False)

# Store results for test
test_results = []

for name, clf in classifiers.items():
    y_test_pred = clf.predict(X_test)
    
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_mean_precision = np.mean(precision_score(y_test, y_test_pred, average=None, zero_division=0))
    test_mean_recall = np.mean(recall_score(y_test, y_test_pred, average=None, zero_division=0))
    test_mean_f1 = np.mean(f1_score(y_test, y_test_pred, average=None, zero_division=0))
    test_min_f1 = np.min(f1_score(y_test, y_test_pred, average=None, zero_division=0))
    
    test_precision_per_class = precision_score(y_test, y_test_pred, average=None, zero_division=0)
    test_recall_per_class = recall_score(y_test, y_test_pred, average=None, zero_division=0)
    test_f1_per_class = f1_score(y_test, y_test_pred, average=None, zero_division=0)
    
    test_results.append({
        'Model': name,
        'Test Accuracy': test_accuracy,
        'Test Mean Precision': test_mean_precision,
        'Test Mean Recall': test_mean_recall,
        'Test Mean F1 Score': test_mean_f1,
        'Test Minimum F1 Score': test_min_f1,
        'Test Precision Per Class': test_precision_per_class.tolist(),
        'Test Recall Per Class': test_recall_per_class.tolist(),
        'Test F1 Score Per Class': test_f1_per_class.tolist()
    })

# Save test results
test_results_df = pd.DataFrame(test_results)
test_results_file = os.path.join(output_dir, 'ablation_ML_model_test_results.csv')
test_results_df.to_csv(test_results_file, index=False)
print("Original Test Results:")
print(test_results_df)

# Sort test results by accuracy (from high to low)
test_results_df_sorted = test_results_df.sort_values(by='Test Accuracy', ascending=False)
sorted_test_results_file = os.path.join(output_dir, 'ablation_sorted_ML_model_test_results.csv')
test_results_df_sorted.to_csv(sorted_test_results_file, index=False)
print("\nSorted Test Results:")
print(test_results_df_sorted)

# Fontsize for classification report and confusion matrix
fontsize = 12

# Display classification report and confusion matrix for each model
for name, clf in classifiers.items():
    y_pred = clf.predict(X_test)

    # Print classification report
    print(f"\nClassification Report for {name}:")
    report = classification_report(y_test, y_pred, target_names=labels)
    print(report)

    # Compute and plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(10, 7))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=labels,
        yticklabels=labels,
        annot_kws={"fontsize": 14}
    )
    plt.title(f'Confusion Matrix for {name}', fontsize=18)
    plt.xlabel('Predicted', fontsize=14)
    plt.ylabel('True', fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()
```
