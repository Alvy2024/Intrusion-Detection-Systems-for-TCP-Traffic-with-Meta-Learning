```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

def divide_by_10e6(x, pos):
    return f'{x * 1e-6:.1f}'

fig, axes = plt.subplots(2, 3, figsize=(30, 20))
axes = axes.flatten()

for i in range(6):
    sns.boxplot(data=train_data.iloc[:, i*3:(i+1)*3], ax=axes[i])
    axes[i].text(0.67, 0.95, 'Values in $10^6$', transform=axes[i].transAxes, ha='center', va='center', 
                 fontsize=22, color='black', weight='bold')
    axes[i].yaxis.set_major_formatter(FuncFormatter(divide_by_10e6))

    x_labels = test_data.iloc[:, i*3:(i+1)*3].columns
    formatted_labels = ['\n'.join(label.split()) for label in x_labels]
    axes[i].set_xticklabels(formatted_labels)

fig.suptitle('Boxplots of Training Data Features', fontsize=24, weight='bold', y=1.02)
plt.tight_layout(rect=[0, 0, 1, 1.02]) 
plt.show()
```
